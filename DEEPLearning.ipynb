# Deep learning
**Deep learning** is basically about tackling complex problems using models called neural networks. These networks mimic the human brain in a way, with multiple layers of artificial neurons that learn from large amounts of data. The deeper the networkâ€”meaning more layersâ€”the better it can recognize patterns or solve intricate tasks!

In deep learning, **classification** uses neural networks to output probability scores for different categories, while **regression** uses neural networks to predict a continuous numeric output.

##  Neuron (Single Unit) - Simple

A **neuron** is a small calculator that:
âœ… takes inputs â†’ multiplies with weights â†’ adds bias â†’ applies activation â†’ gives output

### ðŸ“Œ Neuron Visualization

x1 ----->(w1)---
x2 ----->(w2)--- [ Neuron ] -----> output (y)
x3 ----->(w3)--- +
bias (b)

##  Neuron Formula (Very Important)

### Step 1: Weighted sum
z = (w1*x1 + w2*x2 + w3*x3 + ... + wn*xn) + b

### Step 2: Activation
y = f(z)

### Final neuron output:
y = f(w.x + b)

Weights, Bias, Activation (Super Simple)
## 4) Weights, Bias, Activation (Simple Terms)

### âœ… Weights (w)
- Weights decide **importance** of each input
- Bigger weight = input affects output more

### âœ… Bias (b)
- Bias is an **extra push**
- Helps neuron activate even if inputs are small or zero

### âœ… Activation Function f(z)
- Activation decides **how neuron responds**
- It makes the network learn complex patterns (non-linear learning)

```python
# ==========================================================
# ACTIVATION FUNCTIONS IN DEEP LEARNING (DL)
# ==========================================================
# âœ… What is Activation Function?
# Activation function is a mathematical function applied to a neuron's output.
# It decides "should the neuron activate or not" (how much signal to pass forward).
#
# Neuron formula:
# z = (w1*x1 + w2*x2 + ... + wn*xn) + b
# a = activation(z)
#
# where:
# z = weighted sum (linear output)
# a = activated output (non-linear output)
#
# ==========================================================
# âœ… WHY Activation Function is used? (Main Role)
# ==========================================================
# 1) Adds NON-LINEARITY to the neural network
# 2) Helps model learn complex patterns (curves, shapes, real-world data)
# 3) Controls output range (0 to 1, -1 to 1, etc.)
# 4) Helps backpropagation by providing gradients
#
# ==========================================================
# âœ… What problem does it solve?
# ==========================================================
# Without activation function, a neural network becomes only a linear model.
#
# Example:
# Layer1: z1 = W1x + b1
# Layer2: z2 = W2z1 + b2
#
# If no activation:
# z2 = W2(W1x + b1) + b2
# z2 = (W2W1)x + (W2b1 + b2)
#
# This is STILL a LINEAR equation.
# So even with many layers, the network acts like ONE linear layer.
#
# âœ… So it cannot solve non-linear problems like:
# - XOR problem
# - Image classification
# - Complex decision boundaries
#
# ==========================================================
# âœ… What drawback it overcomes?
# ==========================================================
# Drawback of NO activation:
# âŒ Deep network becomes useless (only linear)
# âŒ Can't learn complex real-world data
# âŒ Can't create curved decision boundaries
#
# Activation function overcomes this by:
# âœ… Making network non-linear (powerful learning)
#
# ==========================================================
# Now we will visualize different activation functions in 2D
# ==========================================================

import numpy as np
import matplotlib.pyplot as plt

# Input values for plotting (x-axis)
x = np.linspace(-10, 10, 1000)

# ==========================================================
# 1) STEP FUNCTION (Binary Activation)
# ==========================================================
# Formula:
# step(x) = 1 if x >= 0 else 0
#
# Used in:
# - Perceptron (old neural networks)
#
# Drawback:
# âŒ Not differentiable -> can't use gradient descent properly
# âŒ Not used in deep learning training

def step_function(x):
    return np.where(x >= 0, 1, 0)

y_step = step_function(x)

plt.figure(figsize=(6,4))
plt.plot(x, y_step)
plt.title("Step Activation Function")
plt.xlabel("x")
plt.ylabel("step(x)")
plt.grid(True)
plt.show()


# ==========================================================
# 2) SIGMOID FUNCTION
# ==========================================================
# Formula:
# sigmoid(x) = 1 / (1 + e^(-x))
#
# Output Range:
# (0 to 1)
#
# Used in:
# - Binary classification output layer
#
# Advantage:
# âœ… Smooth and differentiable
#
# Drawback:
# âŒ Vanishing Gradient Problem:
# When x is very large or very small, gradient becomes near 0
# gradient = sigmoid(x) * (1 - sigmoid(x))
# So learning becomes very slow in deep networks

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

y_sigmoid = sigmoid(x)

plt.figure(figsize=(6,4))
plt.plot(x, y_sigmoid)
plt.title("Sigmoid Activation Function")
plt.xlabel("x")
plt.ylabel("sigmoid(x)")
plt.grid(True)
plt.show()


# ==========================================================
# 3) TANH (Hyperbolic Tangent)
# ==========================================================
# Formula:
# tanh(x) = (e^x - e^(-x)) / (e^x + e^(-x))
#
# Output Range:
# (-1 to 1)
#
# Advantage:
# âœ… Zero-centered (better than sigmoid)
#
# Drawback:
# âŒ Still suffers from vanishing gradient for large values

def tanh(x):
    return np.tanh(x)

y_tanh = tanh(x)

plt.figure(figsize=(6,4))
plt.plot(x, y_tanh)
plt.title("Tanh Activation Function")
plt.xlabel("x")
plt.ylabel("tanh(x)")
plt.grid(True)
plt.show()


# ==========================================================
# 4) ReLU (Rectified Linear Unit)
# ==========================================================
# Formula:
# ReLU(x) = max(0, x)
#
# Output Range:
# (0 to infinity)
#
# Why ReLU is popular?
# âœ… Solves vanishing gradient problem (better than sigmoid/tanh)
# âœ… Faster training
# âœ… Simple computation
#
# Drawback:
# âŒ Dying ReLU problem:
# If neuron output becomes negative always -> gradient becomes 0 -> neuron dies

def relu(x):
    return np.maximum(0, x)

y_relu = relu(x)

plt.figure(figsize=(6,4))
plt.plot(x, y_relu)
plt.title("ReLU Activation Function")
plt.xlabel("x")
plt.ylabel("ReLU(x)")
plt.grid(True)
plt.show()


# ==========================================================
# 5) Leaky ReLU
# ==========================================================
# Formula:
# LeakyReLU(x) = x if x > 0
# LeakyReLU(x) = alpha*x if x <= 0
# (alpha is small like 0.01)
#
# Advantage:
# âœ… Fixes dying ReLU problem (keeps small gradient for negative values)

def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha*x)

y_leaky_relu = leaky_relu(x)

plt.figure(figsize=(6,4))
plt.plot(x, y_leaky_relu)
plt.title("Leaky ReLU Activation Function")
plt.xlabel("x")
plt.ylabel("LeakyReLU(x)")
plt.grid(True)
plt.show()


# ==========================================================
# 6) ELU (Exponential Linear Unit)
# ==========================================================
# Formula:
# ELU(x) = x if x > 0
# ELU(x) = alpha*(e^x - 1) if x <= 0
#
# Advantage:
# âœ… Smooth curve for negative values
# âœ… Reduces dying neuron issue
# âœ… Mean output closer to 0 (helps training)

def elu(x, alpha=1.0):
    return np.where(x > 0, x, alpha*(np.exp(x) - 1))

y_elu = elu(x)

plt.figure(figsize=(6,4))
plt.plot(x, y_elu)
plt.title("ELU Activation Function")
plt.xlabel("x")
plt.ylabel("ELU(x)")
plt.grid(True)
plt.show()


# ==========================================================
# 7) SOFTMAX (Used for Multi-Class Output)
# ==========================================================
# Softmax is not applied on single value, it is applied on a vector.
#
# Formula for each class i:
# softmax(z_i) = e^(z_i) / Î£(e^(z_j))   for all classes j
#
# Output Range:
# (0 to 1) and sum of all outputs = 1
#
# Used in:
# âœ… Multi-class classification output layer (like digits 0-9)

def softmax(z):
    exp_z = np.exp(z - np.max(z))  # subtract max for stability
    return exp_z / np.sum(exp_z)

# Example vector output (logits)
z = np.array([2.0, 1.0, 0.1])
softmax_output = softmax(z)

print("Softmax Example Input (logits):", z)
print("Softmax Output (probabilities):", softmax_output)
print("Sum of probabilities:", np.sum(softmax_output))


# ==========================================================
# âœ… SUMMARY TABLE (Simple)
# ==========================================================
# Step:
# - Old method, not differentiable
#
# Sigmoid:
# - Good for binary output, but vanishing gradient
#
# Tanh:
# - Better than sigmoid, but still vanishing gradient
#
# ReLU:
# - Most used in hidden layers, solves vanishing gradient mostly
# - But dying ReLU problem
#
# Leaky ReLU:
# - Fixes dying ReLU using small negative slope
#
# ELU:
# - Smooth and better negative side behavior
#
# Softmax:
# - Used in output layer for multi-class classification
#
# ==========================================================
# FINAL KEY POINT:
# Activation functions make neural networks powerful by adding NON-LINEARITY.
# Without them, deep learning becomes only linear regression.
# ==========================================================
```
